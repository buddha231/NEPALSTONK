```python
import requests
from datetime import datetime
from bs4 import BeautifulSoup
import json
import pandas as pd
from requests.exceptions import HTTPError
import sys
from datetime import datetime
import calendar
import math
```

```python
url = "http://www.nepalstock.com/gainers"
url2 = "http://www.nepalstock.com/losers"
x = [url, url2]
request_list = list()
for url in x:
    request_list.append(requests.get(url))
df_list = list()
```

```python
count = 1
for request in request_list:
    soup = BeautifulSoup(request.text, 'lxml')
    table = soup.find("tbody").contents       

    table = list(filter(lambda a : a!='\n', table)) #removing the newlines
    table[2:]
    table_list = list()
    for row in table[2:-1]:
        datas = row.find_all("td")
        row = [x.text for x in datas]

        row[1:] = [float(x.replace(",", "")) for x in row[1:] ] #converting quantity rate and amount to float
        ltp = row[1]
        change = row[2]
        previous_close = int(ltp/(1+change/100))
        row.append(previous_close)
        point_change = int(ltp - previous_close)
        row.append(point_change)
        table_list.append(row)

    columns = ["Symbol", "LTP" ,"Change %", "PC", "change"]
    df =  pd.DataFrame(table_list, columns=columns)
    if count == 1:
        df_winner = df
    else:
        df_loser = df
    count+=1
   
    
    
    

```

```python
current_date = datetime.now()
weekday = calendar.day_name[current_date.weekday()]
df_loser.to_csv(f"datas/top_losers_{current_date.date()}_{weekday}")
df_winner.to_csv(f"datas/top_gainers_{current_date.date()}_{weekday}")
```
